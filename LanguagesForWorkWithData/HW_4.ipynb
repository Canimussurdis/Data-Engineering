## Упражнение 3

```python
class Rectangle:
    def __init__(self, lower_left: Point, upper_right: Point):
        self.lower_left = lower_left
        self.upper_right = upper_right
    def area(self):
        width = self.upper_right.x - self.lower_left.x
        height = self.upper_right.y - self.lower_left.y
        return width * height
    
    def perimeter(self):
        width = self.upper_right.x - self.lower_left.x
        height = self.upper_right.y - self.lower_left.y
        return 2 * (width + height)
        
    def contains(self, point: Point):
        return (self.lower_left.x <= point.x <= self.upper_right.x and
                self.lower_left.y <= point.y <= self.upper_right.y)
    
    
p1 = Point(0, 0)
p2 = Point(4, 5)
rect = Rectangle(p1, p2)

point_inside = Point(2, 3)
point_outside = Point(5, 6)

print("Проверка точка(2, 3):", rect.contains(point_inside)) 
print("Проверка точка(5, 6):", rect.contains(point_outside)) 
```

```python
class DecimalCounter:
    def __init__(self, initial_value=0):
        self._counter = initial_value if initial_value >= 0 else 0

    def increment(self):
        self._counter += 1

    def decrement(self):
        if self._counter > 0:
            self._counter -= 1

    def get_counter(self):
        return self._counter
    
    

counter = DecimalCounter()

counter.increment()
print("Current value:", counter.get_counter()) 

counter.decrement()
print("Current value:", counter.get_counter())  

counter.decrement()
print("Current value:", counter.get_counter()) 
```

```python
class Clock:
    def __init__(self, hours = 0, minutes = 0, seconds = 0):
        if not (0 <= hours < 24):
            raise ValueError("Часы должны быть в диапазоне от 0 до 23")
        if not (0 <= minutes < 60):
            raise ValueError("Минуты должны быть в диапазоне от 0 до 59")
        if not (0 <= seconds < 60):
            raise ValueError("Секунды должны быть в диапазоне от 0 до 59")
        
        self.hours = hours
        self.minutes = minutes
        self.seconds = seconds
        
    def add_hour(self):
        self.hours = (self.hours + 1) % 24
        
    def add_minute(self):
        self.minutes += 1
        if self.minutes == 60:
            self.minutes = 0
            self.add_hour()
            
    def add_second(self):
        self.seconds += 1
        if self.seconds == 60:
            self.seconds = 0
            self.add_minute()
    
    def get_time(self):
        return f"{self.hours:02}:{self.minutes:02}:{self.seconds:02}"


clock = Clock(23, 59, 59)

print("Текущее время:", clock.get_time()) 

clock.add_second()
print("Текущее время после добавления 1 секунды:", clock.get_time()) 

clock.add_minute()
print("Текущее время после добавления 1 минуты:", clock.get_time())  

clock.add_hour()
print("Текущее время после добавления 1 часа:", clock.get_time()) 
```

## Упражнение 6

```python
class Clock:
    def __init__(self, hours = 0, minutes = 0, seconds = 0):
        if not (0 <= hours < 24):
            raise ValueError("Часы должны быть в диапазоне от 0 до 23")
        if not (0 <= minutes < 60):
            raise ValueError("Минуты должны быть в диапазоне от 0 до 59")
        if not (0 <= seconds < 60):
            raise ValueError("Секунды должны быть в диапазоне от 0 до 59")
        
        self.hours = hours
        self.minutes = minutes
        self.seconds = seconds
        
    def add_hour():
        self.hour = (self.hour + 1) % 24
        
    def add_minute():
        self.minutes += 1
        if self.minutes == 60:
            self.minutes = 0
            self.add_hour()
            
    def add_second(self):
        self.seconds += 1
        if self.seconds == 60:
            self.seconds = 0
            self.add_minute()
    
    def get_time(self):
        return f"{self.hours:02}:{self.minutes:02}:{self.seconds:02}"

    def __add__(self, other):
        total_seconds = self.seconds + other.seconds
        total_minutes = self.minutes + other.minutes + total_seconds // 60
        total_hours = (self.hours + other.hours + total_minutes // 60) % 24
        
        total_seconds %= 60
        total_minutes %= 60
        
        return Clock(total_hours, total_minutes, total_seconds)

    
clock1 = Clock(10, 59, 30)
clock2 = Clock(3, 45, 40)

print("Время на первых часах:", clock1.get_time())  
print("Время на вторых часах:", clock2.get_time())  

result = clock1 + clock2
print("Результат сложения часов:", result.get_time()) 
```

## Упражнение 7

```python
class Herbivore:
    def __init__(self, hunger = 0):
        self.hunger = hunger
        
    def eat_grass(self, grass:Grass):
        if self.hunger >0:
            self.hunger -= grass.nutrition_value
            self.hunger = max(self.hunger, 0)
            print(f"Животное съело траву с питательностью {grass.nutrition_value}. Уровень голода: {self.hunger}")
        else:
            print(f"Животное не голодное")
                  
class Grass:
    def __init__(self, nutrition_value):
        self.nutrition_value = nutrition_value

                  
                  
low_nutrition_grass = Grass(3)
high_nutrition_grass = Grass(7)

herbivore = Herbivore(hunger = 10)

herbivore.eat_grass(low_nutrition_grass)  

herbivore.eat_grass(high_nutrition_grass) 

herbivore.eat_grass(low_nutrition_grass) 
```

## Упражнение 8

```python
class Water:
    def __add__(self, other):
        if isinstance(other, Air):
            return Storm()
        elif isinstance(other, Fire):
            return Steam()
        elif isinstance(other, Earth):
            return Mud()
        elif isinstance(other, Energy):
            return Life()
        return None

    def __repr__(self):
        return "Water"

class Air:
    def __add__(self, other):
        if isinstance(other, Water):
            return Storm()
        elif isinstance(other, Fire):
            return Lightning()
        elif isinstance(other, Earth):
            return Dust()
        elif isinstance(other, Energy):
            return WindStorm()
        return None

    def __repr__(self):
        return "Air"

class Fire:
    def __add__(self, other):
        if isinstance(other, Water):
            return Steam()
        elif isinstance(other, Air):
            return Lightning()
        elif isinstance(other, Earth):
            return Lava()
        elif isinstance(other, Energy):
            return Plasma()
        return None

    def __repr__(self):
        return "Fire"

class Earth:
    def __add__(self, other):
        if isinstance(other, Water):
            return Mud()
        elif isinstance(other, Air):
            return Dust()
        elif isinstance(other, Fire):
            return Lava()
        elif isinstance(other, Energy):
            return Earthquake()
        return None

    def __repr__(self):
        return "Earth"

class Storm:
    def __repr__(self):
        return "Storm"

class Steam:
    def __repr__(self):
        return "Steam"

class Mud:
    def __repr__(self):
        return "Mud"

class Lightning:
    def __repr__(self):
        return "Lightning"

class Dust:
    def __repr__(self):
        return "Dust"

class Lava:
    def __repr__(self):
        return "Lava"

# доп элементы
class Energy:
    def __add__(self, other):
        if isinstance(other, Water):
            return Life()
        elif isinstance(other, Air):
            return WindStorm()
        elif isinstance(other, Fire):
            return Plasma()
        elif isinstance(other, Earth):
            return Earthquake()
        return None

    def __repr__(self):
        return "Energy"

class Life:
    def __repr__(self):
        return "Life"

class WindStorm:
    def __repr__(self):
        return "Wind Storm"

class Plasma:
    def __repr__(self):
        return "Plasma"

class Earthquake:
    def __repr__(self):
        return "Earthquake"
    
    
    
water = Water()
air = Air()
fire = Fire()
earth = Earth()
energy = Energy()

print(water + air)      
print(water + fire)     
print(water + earth)    
print(air + fire)       
print(air + earth)      
print(fire + earth)     

print(water + energy)      
print(air + energy)        
print(fire + energy)      
print(earth + energy)      
```

## Упражнение 9

```python
class NoMoneyToWithdrawError(Exception):
    def __init__(self, message):
        super().__init__(message)


class PaymentError(Exception):
    def __init__(self, message):
        super().__init__(message)


def print_accounts(accounts):
    """Печать аккаунтов."""
    print("Список клиентов ({}): ".format(len(accounts)))
    for i, (name, value) in enumerate(accounts.items(), start=1):
        print("{}. {} {}".format(i, name, value))


def transfer_money(accounts, account_from, account_to, value):
    """Выполнить перевод 'value' денег со счета 'account_from' на 'account_to'.

    При переводе денежных средств необходимо учитывать:
        - хватает ли денег на счету, с которого осуществляется перевод;
        - перевод состоит из уменьшения баланса первого счета и увеличения
          баланса второго; если хотя бы на одном этапе происходит ошибка,
          аккаунты должны быть приведены в первоначальное состояние
          (механизм транзакции)
          см. https://ru.wikipedia.org/wiki/Транзакция_(информатика).

    Исключения (raise):
        - NoMoneyToWithdrawError: на счету 'account_from'
                                 не хватает денег для перевода;
        - PaymentError: ошибка при переводе.
    """
    if account_from not in accounts or account_to not in accounts:
        raise PaymentError("Один из указанных счетов не существует")

    if value <= 0:
        raise PaymentError("Сумма перевода должна быть положительной")

    if accounts[account_from] < value:
        raise NoMoneyToWithdrawError(f"Недостаточно средств на счете {account_from}")

    initial_from_balance = accounts[account_from]
    initial_to_balance = accounts[account_to]

    try:
        accounts[account_from] -= value
        accounts[account_to] += value
    except Exception as e:
        accounts[account_from] = initial_from_balance
        accounts[account_to] = initial_to_balance
        raise PaymentError(f"Ошибка при переводе: {str(e)}")




if __name__ == "__main__":
    accounts = {
        "Василий Иванов": 100,
        "Екатерина Белых": 1500,
        "Михаил Лермонтов": 400
    }
    print_accounts(accounts)

   
    payment_info = {
        "account_from": "Екатерина Белых",
        "account_to": "Василий Иванов"
    }

    print("Перевод от {account_from} для {account_to}...".
        format(**payment_info))
    try:
        payment_info["value"] = int(input("Сумма = "))
    except ValueError:
        print("Ошибка: введено нечисловое значение суммы.")
        exit()
    
    try:

        transfer_money(accounts, **payment_info)

        print("OK!")
        
    except NoMoneyToWithdrawError as e:
        print(f"Ошибка: {e}")
    except PaymentError as e:
        print(f"Ошибка: {e}")

    print_accounts(accounts)
```

## Упражнение 10

```python
def main():
    total_sum = 0  

    while True:
        user_input = input("Введите число (или нажмите Enter для выхода): ")

        if user_input == "":
            print("Выход из программы.")
            break

        try:
            number = float(user_input)
            total_sum += number
            print(f"Текущая сумма: {total_sum}")
        except ValueError:
            print("Ошибка: введено нечисловое значение. Попробуйте снова.")

if __name__ == "__main__":
    main()
```

```python
letter_to_numeric = {
    'A': 5,
    'B': 4,
    'C': 3,
    'E': 3,
    'F': 2,
    'Р': 2
}

numeric_to_letter = {
    5: 'A',
    4: 'B',
    3: 'C',
    2: 'F'
}

def main():
    while True:
        user_input = input("Введите оценку (буквенную или числовую, или нажмите Enter для выхода): ")

        if user_input == "":
            print("Выход из программы.")
            break

        try:
            numeric_grade = int(user_input)
            if numeric_grade in numeric_to_letter:
                print(f"Числовая оценка {numeric_grade} -> Буквенная оценка: {numeric_to_letter[numeric_grade]}")
            else:
                print("Ошибка: введено недопустимое числовое значение. Попробуйте снова.")
        except ValueError:
            letter_grade = user_input.upper()
            if letter_grade in letter_to_numeric:
                print(f"Буквенная оценка {letter_grade} -> Числовая оценка: {letter_to_numeric[letter_grade]}")
            else:
                print("Ошибка: введено недопустимое буквенное значение. Попробуйте снова.")

if __name__ == "__main__":
    main()
```
