```python
import numpy as np
```

#### Одномерный массив

```python
# Создадим одномерный массив из целых чисел от 0 до 9
array_1d = np.arange(10)
print("Одномерный массив:", array_1d)
```

```python
# Выполняем поэлементное сложение, вычитание, умножение и деление
addition = array_1d + 5          
subtraction = array_1d - 5
multiplication = array_1d * 5
division = array_1d / 5 

print("Суммирование:", addition)
print("Вычитание:", subtraction)
print("Умножение:", multiplication)
print("Деление:", division)
```

```python
# Находим скалярное произведение массива с самим собой
dot_product = np.dot(array_1d, array_1d)

print("Скалярное произведение:", dot_product)
```

```python
# Извлекаем второй и третий элементы из одномерного массива
extracted_elements = array_1d[1:3]
print("Извлеченные элементы:", extracted_elements)
```

```python
# Вычисляем среднее значение, медиану и стандартное отклонение
mean = np.mean(array_1d)
median = np.median(array_1d)
std_deviation = np.std(array_1d)

print("Среднее значение:", mean)
print("Медиана:", median)
print("Стандартное отклонение:", std_deviation)
```

```python
# Изменяем форму массива на двумерный и транспонируем его
array_2d = array_1d.reshape(2, 5)
transposed_array = array_2d.T

print("Двумерный массив:\n", array_2d)
print("Транспонированный массив:\n", transposed_array)
```

#### Двумерный массив

```python
# Создаем двумерный массив размером (3, 3) со случайными целыми числами от 1 до 10
array_2d = np.random.randint(1, 11, size=(3, 3))
print("Двумерный массив:\n", array_2d)
```

```python
# Извлекаем первую строку и последний столбец
first_row = array_2d[0, :]
last_column = array_2d[:, -1]

# Извлекаем элементы, которые больше 5
elements_greater_than_5 = array_2d[array_2d > 5]
# Изменяем значения всех элементов в первом столбце на 0
array_2d[:, 0] = 0
# Извлекаем диагональные элементы
diagonal_elements = np.diag(array_2d)

print("Первая строка:", first_row)
print("Последний столбец:", last_column)
print("Элементы больше 5:", elements_greater_than_5)
print("Диагональные элементы:", diagonal_elements)
```

```python
# Находим максимальные и минимальные значения и их индексы
max_value = np.max(array_2d)
min_value = np.min(array_2d)
max_index = np.unravel_index(np.argmax(array_2d), array_2d.shape)
min_index = np.unravel_index(np.argmin(array_2d), array_2d.shape)

print("Максимальное значение:", max_value, "по индексу:", max_index)
print("Минимальное значение:", min_value, "по индексу:", min_index)
```

```python
# Создаем булеву маску для четных значений
even_mask = (array_2d % 2 == 0)
# Используем маску для извлечения четных элементов
even_elements = array_2d[even_mask]
# Заменяем все элементы, которые меньше 5, на -1
array_2d[array_2d < 5] = -1

print("Четные элементы:", even_elements)
```

```python
def data_head_desc_info(data):
    """
    Функция отображает основную информацию о имеющихся данных из указанного файла.
    Параметры:
    - data (pandas.DataFrame): Входные данные в виде DataFrame.
    Возвращает:
    - None: Функция показывает первые пять строк, описание, 
    графическое отображение количества пропусков и процент пропусков в DataFrame.
    Если произойдет ошибка, будет выведено сообщение об ошибке.
    """
    try:
        print('\033[1m'+'1. Первые пять строк данных'+'\033[0m')
        display(data.head(5))
        print('\033[1m'+'2. Размер датасета'+'\033[0m')
        display(data.shape)
        print('\033[1m'+'3. Типы данных столбцов'+'\033[0m')
        display(data.dtypes)
        print('\033[1m'+'4. Описание данных'+'\033[0m')
        display(data.describe(datetime_is_numeric=True))
        print('\033[1m'+'5. Общая информацию о датафрейме'+'\033[0m')
        display(data.info())
        print('\033[1m'+'6. Графическое отображение процента пропусков'+'\033[0m')
        display((data.isna().mean() * 100).sort_values().to_frame().style.format("{:.2f}", na_rep="")
                .background_gradient('Blues'))
    except:
        return 'Проверьте параметры'
```

```python
def data_isna_nuniq(data):
    """
    Функция для получения короткой информации по определенным столбцам.
    Параметры:
    - data (pandas.DataFrame): Входные данные в виде DataFrame.
    Возвращает:
    - None: Функция показывает данные о пропусках и уникальных значениях в DataFrame.
    Если произойдет ошибка, будет выведено сообщение об ошибке.
    """
    try:
        for column in data.columns:
            print('Столбец:', column)
            print('Количество пропусков:', data[column].isna().sum())
            print('Количество уникальных значений:', data[column].nunique())
    except Exception as e:
        return 'Проверьте параметры: {}'.format(str(e))
```

```python
data_isna_nuniq(df)
```
